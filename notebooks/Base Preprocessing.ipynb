```python

def prep_img(img, base=None):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = ndimage.rotate(img, 2, order=0)[260:550,625:1240,:]
    if base is not None:
        img = cv2.absdiff(base, img)
        img[img < 10] = 0
        img[img > 50] = 255
    return img

img = prep_img(cv2.imread("corners/6.png"), base)


plt.imshow(img);

```
